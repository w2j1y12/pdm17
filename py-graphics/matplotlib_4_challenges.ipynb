> [hint] 검색: "How to add title to subplots in Matplotlib?"

subplot을 이용해서 나란하게 그림(균일분포 좌측/ 정규분포 우측)

```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.hist(np.random.rand(1000))
# 균일분포
```

```
plt.hist(np.random.randn(1000))
# 정규분포
```

### subplot(121), subplot(122)

```
# 1

plt.suptitle("Uniform vs Normal", fontsize=15, fontweight='bold')

plt.subplot(121).set_title("Uniform Distribution")
plt.hist(np.random.rand(1000))
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(122).set_title("Normal Distribution") 
plt.hist(np.random.randn(1000))
plt.xlabel('x')
```

```
# 2

fig = plt.figure()          

ax1 = fig.add_subplot(121)                 
ax1.hist(np.random.rand(1000))               
ax1.title.set_text('Uniform Distribution')   
ax1.set(xlabel = 'x', ylabel = 'y')  

ax2 = fig.add_subplot(122)                  
ax2.hist(np.random.randn(1000))              
ax2.title.set_text('Normal Distribution')  
ax2.set(xlabel = 'x')

fig.tight_layout()    
plt.show()
```

### [도전 2.] 0~9 사이의 무작위 정수 1000개를 만들어서 각 수의 빈도에 대한 히스토그램, 막대그래프, 꺽은선 그래프를 그리고 비교하시오.
- subplot 이용
- style(제목, x-축 레이블, y-축 레이블, ...)



```
n1000 = np.random.randint(0,10,1000)
```

```
value, count= np.unique(n1000, return_counts=True)
# count
```

```
# 히스토그램

plt.hist(n1000)
plt.show()
```

```
# 막대그래프

plt.bar(value,count) #,color="bgrcmyk")  # 'bgrcmykw
```

```
# 꺽은선 그래프

plt.plot(value,count,marker='o')
plt.show()
```

```
#1

plt.suptitle("Histogram vs Bar vs Line", fontsize=15, fontweight='bold')

plt.subplot(131).set_title("Histogram")
plt.hist(n1000)
plt.xlabel('value')
plt.ylabel('count')

plt.subplot(132).set_title("Bar")
plt.bar(value,count)
plt.xlabel('value')

plt.subplot(133).set_title("Line")
plt.plot(value,count,marker='o')
plt.xlabel('value')
```

```
# 2

fig = plt.figure()           

ax1 = fig.add_subplot(131)                 
ax1.hist(n1000)                           
ax1.title.set_text('Histogram')            
ax1.set(xlabel = 'value', ylabel = 'count')

ax2 = fig.add_subplot(132)                  
ax2.bar(value,count)                       
ax2.title.set_text('Bar graph')             
ax2.set(xlabel = 'value') 

ax3 = fig.add_subplot(133)           
ax3.plot(value,count,marker='o')                     
ax3.title.set_text('Line graph')            
ax3.set(xlabel = 'value')

fig.tight_layout() 
plt.show()
```
