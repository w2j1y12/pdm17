```python
# sin그래프를 그린다
s = np.sin(np.pi*np.arange(0.0, 2.0, 0.01))
plt.plot(s)    # sin의 그래프를 그림
# 200개의 점으로 그려진 sin 함수
# 근데 x를 저렇게 지정했기 때문에 x축이 0-200
```

```python
s1 = np.sin(np.pi*np.arange(0.0, 2.0, 0.1))
plt.plot(np.arange(0.0, 2.0, 0.1), s1, '*', ms = 10)
# marker = 'o' .../ ms = marker size
```

## 산포도 (scatter plot)

```python
# 산포도(scatter plot)를 그린다
# 마커를 사용하여 산포도를 작성
# 표준 정규 분포에 따라 난수 생성
x = np.random.randn(5000)    
y = np.random.randn(5000)
plt.figure(figsize=(5,5))
t = plt.plot(x, y, 'o', alpha=0.9)   #, ms=8)

# 가우시안 분포라서 평균 0, 표준편차 1
# 적절한 알파 값으로 시각화를 표현을 잘 해야함
```

### [도전] 균일분포(백색잡음)의 산포도 그리기

```python
# 균일 분포에 따라 난수 생성하고 산포도 그리기

x = np.random.rand(5000)    
y = np.random.rand(5000)
plt.figure(figsize=(5,5))
t = plt.plot(x, y, 'o', alpha=0.1)   #, ms=8)

# randn -> rand
# 거의 균일하게 0에서 1사이 값이 나옴
```

### 막대 그래프 (Bar graph)

```python
# 막대그래프(bar graph)를 그린다.
means = [1, 2, 3] # 평균
stddevs = [0.2, 0.4, 0.5] # 표준편차
bar_labels = ['bar 1', 'bar 2', 'bar 3']

# plot bars
x_pos = list(range(len(bar_labels)))
# range(len(bar_labels)) = range(3) = (0,2)
plt.bar(x_pos, means, yerr=stddevs, color=('r','g','b'))

plt.show()
```

### 히스토그램 (histogram)

```python
# 히스토그램 (histogram)
rng = np.random.RandomState(123)  # different from random.seed()
x = rng.normal(0, 20, 1000) 
# randomstate: seed를 만들어서 제너레이터를 만듦(seed랑은 다름)
# rng = random number generator
# normal을 하면 randomstate로 만든 랜덤 정규분포를 (평균, 표준편차, 개수)

# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size
# -100부터 100까지 5단위로 -> 막대의 개수 40개
plt.hist(x, bins=bins)
# 고정간격
plt.show()
```

### [도전하기] 위의 histogram을 정수 bins를 지정해서 그리시오.

```python
# 히스토그램 (histogram)
rng = np.random.RandomState(123)  # different from random.seed()
x = rng.normal(0, 20, 1000) 

# fixed bin size
# bins = np.arange(-100, 100, 5) # fixed bin size
bins=100
# bins를 더 늘리면 더 촘촘하게 만들어짐

plt.hist(x, bins=bins)
plt.show()
```

```python
# 중첩 히스토그램
rng = np.random.RandomState(123)
x1 = rng.normal(0, 20, 1000) 
x2 = rng.normal(20, 10, 1000)
```

```python
x1.shape, x2.shape
```

```python
# fixed bin size
bins = np.arange(-100, 100, 5) # fixed bin size

plt.hist(x1, bins=bins, alpha=0.5)
plt.hist(x2, bins=bins, alpha=0.5)
plt.show()
# 평균 부근에서 값이 크고, 표준편차가 작을 수록 높게? 가파르게? 생김
```
